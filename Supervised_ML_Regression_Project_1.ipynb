# **Project Summary -**

In this project, we're using data to predict how many bikes people will rent in Seoul. We looked at things like the date, time, weather, and temperature to figure this out.

We cleaned up the data, making sure there were no mistakes or missing values. We also removed some data that didn't seem important.

Then, we studied the data and found some interesting things:

* People like renting bikes in all seasons.

* They rent more when it's not a holiday.

* Most rentals happen during work hours.

* Biking is popular all year round.

* Rentals increased a lot from 2017 to 2018.

* People don't rent bikes much at night.

* More rentals happen during office commute times.

* People often rent bikes in the evening.

* Humidity and wind speed affect rentals.

* Rentals go up when it's not raining.

* Snow doesn't stop rentals until it's really heavy.

* The first and last days of the month are busy.

* Rentals peak in June and October.

* Visibility doesn't change rentals much.

* Fewer people rent bikes in winter.

* People still rent bikes on holidays, but less.

* Most rentals are during "functional hours."

We also verified the presumptions of the regression machine learning model. We restore the distribution of the features Rented Bike Count, Wind speed (m/s), Solar Radiation (MJ/m2), Visibility (10 m), Rainfall (mm), and Snowfall (cm) to normal using log and square root transformation. We also check to see if there is a linear relationship between the independent and dependent variables.

Furthermore, we use a heatmap and the variance inflation factor to examine multicollinearity in independent variables (VIF). Due to their high VIFs, we eliminated the variables year, dew point temperature (°C), and humidity (%). Once categorical variables like seasons, holidays, and working days were encoded, our dataset was ready to be used with a machine-learning model.

To make it easier for a model to learn and understand the problem, we divided the dataset into dependent and independent features and scaled them to the same length.

Then we implement machine learning models like linear regression, Lasso (L1), Ridge (L2), ElasticNet, Decision Tree regressors, Random Forest, and XGBoost regression.We also made a smart machine that can predict bike rentals. We trained it with our data and used a model called XGBoost.  We obtained a high accuracy of 0.9026 and a low root mean squared error of 3.80 from the XGBoost model, become final optimal model for prediction.It's really accurate

The Functioning Days, Rainfall (mm), and Seasons variables have a significant influence on the XBoost model, according to our plot of feature significance. Additionally, we explain our XGBoost model using SHAP.
